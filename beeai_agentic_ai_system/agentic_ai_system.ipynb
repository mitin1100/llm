```python
import asyncio
import logging
from beeai_framework.backend import ChatModel, ChatModelParameters, UserMessage, SystemMessage

# Cấu hình logging để dễ theo dõi lỗi nếu có
logging.basicConfig(level=logging.INFO)

# async def main():
    # SỬA TẠI ĐÂY: Dùng ChatModel.from_name thay vì gọi trực tiếp ChatModel
    # Định dạng chuẩn thường là "provider:model_name" (Ví dụ: "google:gemini-2.5-flash")
llm = ChatModel.from_name(
    "google:gemini-2.5-flash", 
    ChatModelParameters(temperature=0)
)

messages = [
    SystemMessage(content="You are a helpful AI assistant and creative writing expert."),
    UserMessage(content="Help me brainstorm a unique business idea for a food delivery service that doesn't exist yet.")
]

# SỬA TẠI ĐÂY: Dùng `await llm.create` và truyền vào một Dictionary chứa key "messages"
response = llm.create({"messages": messages})

# In kết quả trả về từ mô hình
print(response.get_text_content())

# if __name__ == "__main__":
#     asyncio.run(main())
```
